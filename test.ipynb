```python
from spatialMCMC import SpatialRegression, DataGrid
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')

y_test = df["bicycle.theft"].to_numpy()
X_test = df[['xi', 'yi']].to_numpy()
```

```python
np.random.seed(1)
```

```python
factor = 3
N = X_test.shape[0]
idx = np.random.choice(N, size=int(np.ceil(N / factor)), replace=False)
X = X_test[idx, :]
y = y_test[idx]
```

```python
model = SpatialRegression()
```

```python
test_grid = DataGrid(coords=X_test)
```

```python
test_grid = DataGrid(grid_limits=[[0,20],[0,20]], grid_ns=[20,20])
```

```python
predictions = model.fit_and_predict(X, y, test_grid) #maybe remove X from X_test first?
```
